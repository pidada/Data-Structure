`````python
# Python

list2 = [28, 1, 7, 4, 37]
length = len(list2)
# 
for i in range(length-1):
    # iii
    for j in range(length-1-i):
        if list2[j] > list2[j+1]:
            list2[j], list2[j+1] = list2[j+1], list2[j]
print(list2)
`````

`````python
# 
def bubble_sort(list1):
    length = len(list1)
    for i in range(length-1):
        for j in range(length-1-i):
            if list1[j] > list1[j+1]:
                list1[j], list1[j+1] = list1[j+1], list1[j]
    return list1


if __name__ == "__main__":
    list1 = [28, 1, 7, 4, 37]
    bubble_sort(list1)
    print(list1)
    
`````
