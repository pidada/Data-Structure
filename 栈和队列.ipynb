### 栈
栈也是一种数据呈`线性排列`的数据结构。用户只能访问最新的数据。

栗子：
我们栈看成是一摞书，当我们放新书的时候，只能放在最上面；取书的时候，只能取出最上面的书。
#### 入栈和出栈
- 入栈：往栈中添加新的数据，添加位置在栈顶
- 出栈：从栈中取出新的数据，只能取出栈顶的数据
- 看图中的下端是封闭，不能进行操作



#### LIFO
在栈中，最后进去的数据（在栈顶）最先被取出来（只能从栈顶开始），叫做`Last In First Out`，简称`LIFO`

#### 特点
- 数据是线性排列
- 添加和删除操作只能在一端进行，即栈顶
- 想要删除中间位置的数据，只能把数据移动到栈顶，再进行操作
- 联想Python中的pop和append操作
    - pop：删除末尾的元素
    - append：末尾追加元素
    
    
### 队列
>队列也是一种线性排列的数据，添加和删除数据的操作能够在两端进行。将队列比喻成一个队伍：新来的人只能在队尾，处理总是从队首开始。
#### 入队和出队
- 入队：往队列中添加数据
- 出队：删除队列中的数据，删除的数据是最先进入的数据
#### FIFO
在队列中，最先进入队列的数据最先被删除，即先进先出`First In First Out`，简称为`FIFO`。
**如果想操作中间的数据，必须通过出队操作将数据变成首位后才能进行操作。**
