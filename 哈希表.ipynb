### 哈希表
#### 哈希基础
>哈希表（`Hash table`）或者散列表，是根据关键码值或者键值对(`Key-value`)而直接进行访问的数据结构。它通过把关键码值映射到表中一个位置来访问记录，以加快查找的速度。这个映射函数叫做`散列函数`，存放记录的数组叫做`散列表`。
#### 特点
- 类似`Python`中的字典，根据`key-value`存储数据
- 键：数据的标识符；值：数据的内容
#### 存储和访问数据
- 利用哈希函数计算数据的键，也就是哈希值
- 将哈希值/数据长度，求余数即`mod`运算
- 将余数作为编号，将数据存入该编号中
- 如果遇到余数相同的情况，发生冲突，使用链表，在已有数据的后面继续存储新的数据
- 访问数据也是类似的操作
#### 注意点
- 利用哈希函数快速访问数组中的数据
- 若数组的空间太小，使用哈希表容易发生冲突，线性查找的频率也会更高
- 空间太大，会造成内存空间的浪费
- 解决哈希冲突的方法
    - 链表地址法（上面提及）
    - 开放地址法：发生冲突时，立即计算出一个候补地址，将数据存放进去
