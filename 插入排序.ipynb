### 插入排序
插入排序是一种从序列左端开始对数据进行排序的算法。
- 左侧：已经排好序的数据
- 右侧：需要进行排序的数据
- 从右侧未排序的区域中选取一个数，插入左边排好序的合适位置上



```python
# Python代码实现

list1 = [1, 9, 4, 7, 2, 8, 3]
n = len(list1)

for i in range(n-1):
    value = list1[i]
    for j in range(0, i):
        if list1[j] > value:
            list1[j], list1[j+1] = list1[j+1], list1[j]
print(list1)
```
