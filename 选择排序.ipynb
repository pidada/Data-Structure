### 选择排序
#### 算法思想
>选择排序的思想是从待排序的数据中选出`最小`的值，与其`最左边`的数字进行交换，重复以上步骤。在序列中查找最小值使用的是**线性查找**。
- 1、先用list[0]和list[1]~list[len(list)]比较，如果list[0]大，交换位置，将较小值放在list[0]即首位上。
- 2、再用list[1]和list[2]~list[len(list)]比较，如果list[1]大，交换位置，放在未排序的序列首位上。
- 3、重复1和2的过程。



```python
# python 实现

def select_sort(list1):
    n = len(list1)
    # i外层控制轮数，总长度为n，需要循环(n-1)轮，最后一个数无需比较；同时i也表示
    for i in range(n-1):
        # 假定当前最小值的索引是i
        min_index = i
        # 当第j轮开始比较的时候，其实是要和后面的(n-i-1)个元素进行比较
        for j in range(i+1, n):   
            # 如果本轮中我们假定的最小值，也就是索引为min_index处的值比后面未排序的元素要大，说明 min_index处的值不是最小的
            if list1[min_index] > list1[j]:  
                #  找到最小值元素的索引  
                min_index = j   
        # 退出整个循环，交换我们设定的最小值list1[i]和实际找到的最小值list1[min_index]的值，也就是上步中交换得到的list1[j]的值
        list1[i], list1[min_index] = list1[min_index], list1[i]    
    return list1
    
if __name__ == "__main__":
    list1 = [1,9,3,5,2,8,6]
    print("排序前",list1)
    select_sort(list1)
    print("排序后",list1)
```

```python
list1 =  [1, 9, 3, 5, 2, 8, 6]
length = len(list1)

# 外循环控制总次数
for i in range(length-1):
    # 内循环，假定索引为i的最小，它和后面的[i+1, length]的元素进行比较
    for j in range(i+1, length):
        if list1[i] > list1[j]:
            # 如果不满足list1[i]最小，交换位置
            list1[i], list1[j] = list1[j], list1[i]
print(list1)
        
```
